# Dataframe

```python
def get_forecast(response, i):
  date = response["forecast"]["forecastday"][0]["hour"][i]["time"].split()[0]
  hour = int(response["forecast"]["forecastday"][0]["hour"][i]["time"].split()[1].split(":")[0])
  condition = response["forecast"]["forecastday"][0]["hour"][i]["condition"]["text"]
  temp_c = response["forecast"]["forecastday"][0]["hour"][i]["temp_c"]
  rain = response["forecast"]["forecastday"][0]["hour"][i]["will_it_rain"]
  change_of_rain = response["forecast"]["forecastday"][0]["hour"][i]["chance_of_rain"]

  return date, hour, condition, temp_c, rain, change_of_rain
```

```python
data = []

number_hours = len(response["forecast"]["forecastday"][0]["hour"])

for i in tqdm(range(number_hours)):
  data.append(get_forecast(response, i))
```

```python
data
```

```python
columns = ["date", "hour", "condition", "temp_c", "rain", "change_of_rain"]

df = pd.DataFrame(data, columns=columns)

df
```

```python
df_rain = df[(df["rain"] == 1) & (df["hour"] > 6) & (df["hour"] < 22)]
df_rain = df_rain[["hour", "condition"]]
df_rain.set_index("hour", inplace = True)

df_rain
```

# Template message

```python
text_message = f"\nHi! \n\n\n Today's weather forecast {df['date'][0]} in {country_query} is: \n\n\n {str(df_rain)}"

text_message
```

# Send text message from Twilio

```python
client = Client(account_sid, auth_token)

message = client.messages.create(
    to="+000000000",
    from_="+000000000",
    body=text_message)

print(message.sid)
```
